```python
import pandas as pd

# Carregar o arquivo aps o upload
df = pd.read_csv("/content/listings.csv")

# Exibir as primeiras linhas do dataset
df.head()
df.columns
# Importar bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Converter last_review para o formato de data
df["last_review"] = pd.to_datetime(df["last_review"])

# Converter price para nmero (removendo smbolos como "$" se houver)
df["price"] = df["price"].astype(float)

# Criar uma nova coluna apenas com o ms
df["month"] = df["last_review"].dt.month
df.head()
# Criando um grfico de linha para visualizar a variao dos preos mdios ao longo dos meses
plt.figure(figsize=(10,5))
sns.lineplot(x=df["month"], y=df["price"], marker="o", ci=None , color='hotpink', )

# Melhorando os rtulos do grfico
plt.xticks(range(1,13), ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"])
plt.xlabel("Ms")
plt.ylabel("Preo Mdio (R$)")
plt.title("Variao do Preo do Airbnb ao Longo do Ano")
plt.show()

```

```python
# Preo mdio por bairro
preco_bairro = df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)

# Nmero de avaliaes por bairro
avaliacoes_bairro = df.groupby('neighbourhood')['number_of_reviews'].sum().sort_values(ascending=False)

# Exibindo os 10 bairros com maior preo mdio
print(preco_bairro.head(10))
# Selecionando os 10 bairros com maior preo mdio
top10_preco = preco_bairro.head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top10_preco.values, y=top10_preco.index, palette="magma")
plt.xlabel("Preo Mdio (R$)")
plt.ylabel("Bairro")
plt.title("Top 10 Bairros com Maior Preo Mdio")
plt.show()

# Exibindo os 10 bairros com maior nmero de avaliaes
print(avaliacoes_bairro.head(10))
# Selecionando os 10 bairros com maior nmero de avaliaes
top10_avaliacoes = avaliacoes_bairro.head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top10_avaliacoes.values, y=top10_avaliacoes.index, palette="viridis")
plt.xlabel("Nmero de Avaliaes")
plt.ylabel("Bairro")
plt.title("Top 10 Bairros com Maior Nmero de Avaliaes")
plt.show()



```

```python
# Calculando a correlao entre preo e nmero de avaliaes
correlacao = df[['price', 'number_of_reviews']].corr()

# Exibindo a correlao
print(correlacao)
```
