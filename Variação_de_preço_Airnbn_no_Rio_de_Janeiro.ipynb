```python
import pandas as pd

# Carregar o arquivo após o upload
df = pd.read_csv("/content/listings.csv")

# Exibir as primeiras linhas do dataset
df.head()
df.columns
# Importar bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Converter last_review para o formato de data
df["last_review"] = pd.to_datetime(df["last_review"])

# Converter price para número (removendo símbolos como "$" se houver)
df["price"] = df["price"].astype(float)

# Criar uma nova coluna apenas com o mês
df["month"] = df["last_review"].dt.month
df.head()
# Criando um gráfico de linha para visualizar a variação dos preços médios ao longo dos meses
plt.figure(figsize=(10,5))
sns.lineplot(x=df["month"], y=df["price"], marker="o", ci=None , color='hotpink', )

# Melhorando os rótulos do gráfico
plt.xticks(range(1,13), ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"])
plt.xlabel("Mês")
plt.ylabel("Preço Médio (R$)")
plt.title("Variação do Preço do Airbnb ao Longo do Ano")
plt.show()

```

```python
# Preço médio por bairro
preco_bairro = df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)

# Número de avaliações por bairro
avaliacoes_bairro = df.groupby('neighbourhood')['number_of_reviews'].sum().sort_values(ascending=False)

# Exibindo os 10 bairros com maior preço médio
print(preco_bairro.head(10))
# Selecionando os 10 bairros com maior preço médio
top10_preco = preco_bairro.head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top10_preco.values, y=top10_preco.index, palette="magma")
plt.xlabel("Preço Médio (R$)")
plt.ylabel("Bairro")
plt.title("Top 10 Bairros com Maior Preço Médio")
plt.show()

# Exibindo os 10 bairros com maior número de avaliações
print(avaliacoes_bairro.head(10))
# Selecionando os 10 bairros com maior número de avaliações
top10_avaliacoes = avaliacoes_bairro.head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top10_avaliacoes.values, y=top10_avaliacoes.index, palette="viridis")
plt.xlabel("Número de Avaliações")
plt.ylabel("Bairro")
plt.title("Top 10 Bairros com Maior Número de Avaliações")
plt.show()



```

```python
# Calculando a correlação entre preço e número de avaliações
correlacao = df[['price', 'number_of_reviews']].corr()

# Exibindo a correlação
print(correlacao)
```
